# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()
    if "calculate" in query_lower:
        return {
            "type": "calculation",
            "result": calculator(query_lower.replace("calculate","").strip())
        }
    elif "keywords" in query_lower:
        return {
            "type": "keywords",
            "result": extract_keywords(query_lower.replace("keywords","").strip())
        }
    else:
        return {
            "type": "general",
            "result": "I can calculate expressions, extract keywords, or do general queries."
        }
    return {
          "type": "unknown",
          "result": "Not implemented"
    }
````

## 📦 Expected Output Format

```

  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# ---------------- BONUS SECTION ----------------

def log(message):
    print("[LOG]", message)

def word_count(text):
    return len(text.split())

def agent(query):
    try:
        query = query.strip()
        if query == "":
            return {"type":"error","result":"Empty input provided."}

        q = query.lower()

        # Calculator routing
        if ("calculate" in q) or any(op in q for op in ["+","-","*","/"]):
            log("Calculator tool selected")
            expr = query.lower().replace("calculate","").strip()
            return {"type":"calculation","result":calculator(expr)}

        # Keyword routing
        elif ("keywords" in q) or ("extract" in q):
            log("Keyword extractor selected")
            text = query.replace("keywords","").replace("extract","").strip()
            return {"type":"keywords","result":extract_keywords(text)}

        # Extra tool
        elif "count words" in q:
            log("Word count tool selected")
            text = query.lower().replace("count words","").strip()
            return {"type":"word_count","result":word_count(text)}

        else:
            log("Fallback response")
            return {"type":"general","result":"I can calculate expressions, extract keywords, or count words."}

    except Exception as e:
        return {"type":"error","result":str(e)}

# Validation
tests = [
    "calculate 12*8",
    "extract keywords Artificial Intelligence is changing healthcare",
    "count words ChatGPT helps students learn quickly",
    "Hello Agent",
    ""
]

for t in tests:
    print(agent(t))

# Interactive loop
while True:
    user = input("Enter query (type exit to stop): ")
    if user.lower()=="exit":
        break
    print(agent(user))
````
